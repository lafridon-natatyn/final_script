4. *Modelling*

```python
columns_for_drop = ['session_id','utm_medium_clean', 'device_os_clean', 'device_brand_clean', 'device_browser_clean', 'geo_country_clean','utm_source_clean','utm_adcontent_clean','geo_city_clean','device_category']
df_prepared = df_clean.drop(columns_for_drop, axis= 1)
df_prepared
```

```python
df_clean.info()
```

```python
df_prepared.info()
```

```python
df_prepared['target'].value_counts()
```

```python
df_prepared.shape
```

```python
x = df_prepared.drop(['target'], axis=1)
y = df_prepared['target'] 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
logreg = LogisticRegression(class_weight='balanced')
```

```python
logreg.fit(x_train, y_train)
```

```python
logreg_pred_train = logreg.predict(x_train) 
logreg_pred = logreg.predict(x_test)
print(accuracy_score(y_train, logreg_pred_train))
print(accuracy_score(y_test, logreg_pred))
```

```python
confusion_matrix(y_test, logreg_pred)
```

```python
roc_auc_score(y, logreg.decision_function(x))
```

```python
coef_list = list(zip(x_train.columns, logreg.coef_))
```

```python
mlp = MLPClassifier(random_state=42, max_iter=500, hidden_layer_sizes=(100,20), activation = 'tanh')
mlp.fit(x_train, y_train)
```

```python
mlp_pred_train = mlp.predict(x_train) 
mlp_pred = mlp.predict(x_test)
print(accuracy_score(y_train, mlp_pred_train))
print(accuracy_score(y_test, mlp_pred))
```

```python
confusion_matrix(y_test, mlp_pred)
```

```python
roc_auc_score(y, mlp.predict_proba(x)[:, 1])
```

```python
coef_list = list(zip(x_train.columns, mlp.coef_))
```

```python
rf_clf = RandomForestClassifier(random_state=42)
```

```python
rf_clf.fit(x_train, y_train)
```

```python
predicted_train_rf = rf_clf.predict(x_train)
predicted_test_rf = rf_clf.predict(x_test)
```

```python
print(accuracy_score(y_train, predicted_train_rf))
print(accuracy_score(y_test, predicted_test_rf))
```

```python
confusion_matrix(y_test, predicted_test_rf)
```

```python
roc_auc_score(y, rf_clf.predict_proba(x)[:, 1])
```

```python
coef_list = list(zip(x_train.columns, rf_clf.coef_))
```

По итогам моделирования лучшей моделю оказалась модель случайного леса
